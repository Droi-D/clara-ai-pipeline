## Cell 4 — Pipeline A: Demo Calls → v1 Outputs
Processes all 5 demo transcripts. Each makes 2 Groq API calls (extraction + spec generation).

```python
# Cell 4 — Pipeline A: Demo transcripts → v1 outputs (account_memo + agent_spec)
import sys, importlib

# Force fresh module imports so changes in Cell 2 are picked up
for mod in list(sys.modules.keys()):
    if mod in ("utils", "prompts", "pipeline_a", "pipeline_b"):
        del sys.modules[mod]

from pipeline_a import run_pipeline_a

TRANSCRIPTS_DIR = "sample_transcripts"
results = []

print("📞 PHASE 1 — Demo Transcripts (Pipeline A → v1 outputs)")
print("-" * 58)

for i in range(1, 6):
    account_id      = f"ACC_{i:03d}"
    transcript_path = f"{TRANSCRIPTS_DIR}/demo_{i}.txt"
    result = {
        "account_id":    account_id,
        "demo_ok":       False,
        "onboarding_ok": False,
        "v1_saved":      False,
        "v2_saved":      False,
    }
    results.append(result)
    try:
        run_pipeline_a(transcript_path, account_id)
        result["demo_ok"]  = True
        result["v1_saved"] = True
    except Exception as exc:
        print(f"  ❌ Pipeline A failed for {account_id}: {exc}")

print("\n✅ Pipeline A complete!")
```

## Cell 5 — Pipeline B: Onboarding Calls → v2 Outputs
Processes all 5 onboarding transcripts. Each makes 2 Groq API calls (update extraction + spec regeneration).

```python
# Cell 5 — Pipeline B: Onboarding transcripts → v2 outputs (updated memo + spec + changelog)
import sys

for mod in list(sys.modules.keys()):
    if mod in ("utils", "prompts", "pipeline_b"):
        del sys.modules[mod]

from pipeline_b import run_pipeline_b

TRANSCRIPTS_DIR = "sample_transcripts"

print("📋 PHASE 2 — Onboarding Transcripts (Pipeline B → v2 outputs)")
print("-" * 62)

for i in range(1, 6):
    account_id      = f"ACC_{i:03d}"
    transcript_path = f"{TRANSCRIPTS_DIR}/onboarding_{i}.txt"
    result = next(r for r in results if r["account_id"] == account_id)

    if not result["v1_saved"]:
        print(f"  ⏭️  Skipping {account_id} — v1 not saved (Pipeline A failed)")
        continue

    try:
        run_pipeline_b(transcript_path, account_id)
        result["onboarding_ok"] = True
        result["v2_saved"]      = True
    except Exception as exc:
        print(f"  ❌ Pipeline B failed for {account_id}: {exc}")

print("\n✅ Pipeline B complete!")
```

## Cell 6 — Final Summary
Prints the status table, full file tree, and a sample changelog preview.

```python
# Cell 6 — Final summary of all generated outputs
import os, json

# ── Summary table ──────────────────────────────────────────────────────────
print("=" * 68)
print(f"  {'Account':<10} {'Demo':^8} {'Onboarding':^12} {'v1 Saved':^10} {'v2 Saved':^10}")
print("=" * 68)
for r in results:
    print(
        f"  {r['account_id']:<10} "
        f"{'✅' if r['demo_ok']       else '❌':^8} "
        f"{'✅' if r['onboarding_ok'] else '❌':^12} "
        f"{'✅' if r['v1_saved']      else '❌':^10} "
        f"{'✅' if r['v2_saved']      else '❌':^10}"
    )
print("=" * 68)

# ── File tree ──────────────────────────────────────────────────────────────
print("\n📁 Generated output files:")
print("-" * 58)
for acc_id in [f"ACC_{i:03d}" for i in range(1, 6)]:
    base_dir = f"outputs/accounts/{acc_id}"
    if not os.path.exists(base_dir):
        continue
    print(f"\n  {acc_id}/")
    for version in ["v1", "v2"]:
        vdir = f"{base_dir}/{version}"
        if os.path.exists(vdir):
            for fname in sorted(os.listdir(vdir)):
                fpath = f"{vdir}/{fname}"
                size  = os.path.getsize(fpath)
                print(f"    {version}/{fname:<30} {size:>8,} bytes")

# ── Changelog preview for ACC_001 ─────────────────────────────────────────
cl_path = "outputs/accounts/ACC_001/v2/changelog.json"
if os.path.exists(cl_path):
    print("\n📝 Sample Changelog (ACC_001 v1 → v2):")
    print("-" * 58)
    cl = json.load(open(cl_path))
    print(f"  Account : {cl.get('account_id')}")
    print(f"  Updated : {cl.get('updated_at')}")
    for ch in cl.get("changes", []):
        print(f"  • {ch.get('field')}")
        print(f"      old → {ch.get('old_value')}")
        print(f"      new → {ch.get('new_value')}")
        print(f"      why : {ch.get('reason')}")
```
